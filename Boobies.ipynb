````python
import requests
import pandas as pd
````

````python
# Grab the README file which contains the data dictionary for the dataset

dictionary_url = 'http://peipa.essex.ac.uk/pix/mias/README'
print(dictionary_url)

boob_dictionary = requests.get(dictionary_url)

clean_boobs = []

for line in boob_dictionary.text.split('\n'):
  if (line[:3] == 'mdb'):
      boob_fields = line.split(' ')
      clean_boobs.append(boob_fields)

dictionary_df = pd.DataFrame(clean_boobs)

dictionary_df.columns = ['ref','tissue_type_code','abnormality_class_code','severity_code','x','y','radius']

dictionary_df.head()

````

````python
# add the categorical names in
# create the lookups which can be

tissue_type_label = {'F':'fatty','G':'fatty-glandular','D':'dense-glandular'}

tissue_type_label_df = pd.DataFrame(tissue_type_label,columns=['s','t'])

tissue_type_label_df.head()

abnormality_class_label = {
     'CALC': 'Calcification',
     'CIRC': 'Well-defined/circumscribed masses',
     'SPIC': 'Spiculated masses',
     'MISC': 'Other, ill-defined masses',
     ''    : 'unknown'
}

severity_label = {
      'B': 'benign',
      'M': 'malignant',
      '': 'unknown'
 }
````

````python
## join the dictionary dataframe with the human text labels so its easier to read later

dictionary_df['tissue_type'] = dictionary_df['tissue_type_code'].map(tissue_type_label)
dictionary_df['abnormality_class'] = dictionary_df['abnormality_class_code'].map(abnormality_class_label)
dictionary_df['severity'] = dictionary_df['severity_code'].map(severity_label)

dictionary_df.head()
````
